### Problem 1 (Warmup)

```julia
using JuMP, Clp

m = Model(solver = ClpSolver())

# Our model in detail
@variable(m, 0 <= x_1 <= 3)                # Define x_1
@variable(m, 0 <= x_2 <= 3)                # Define x_2
@constraint(m, 2x_1 + x_2 <= 5)            # Constraint 1
@constraint(m, x_1 + x_2 >= 2)             # Constraint 2
@objective(m, Max, 5x_1 - x_2)             # Maximize

# 1b/1c
status = solve(m)
println(status)
println("Value of x_1: ", getvalue(x_1))
println("Value of x_2: ", getvalue(x_2))
println("Optimal (max) objective value: ", getobjectivevalue(m))

```

1a. I am using the Clp solver since it is specialized and fast at solving linear programs,
which is what we are working with here. ECOS and SCS, as seen from the "Introduction 2" lecture slides, 
are more generalized and therefore slower.

1b/c. See code comments

1d. If you change constraint 1 to have  on the right hand side, the model solution becomes infeasible (not optimal).
This makes sense since constaint 2 dictates that x_1 + x_2 must be >= 2. These two constraints cannot both be
true at the same time.

### Problem 2

#### Model Without Separating Parameters (2a)

```julia
using JuMP, Clp

m = Model(solver = ClpSolver())

@variable(m, w >= 0) # Number of acres of wheat to plant
@variable(m, c >= 0) # Number of acres of corn to plant

@constraint(m, 3w + 2c <= 100) # Limited number of workers (100)
@constraint(m, 2w + 4c <= 130)  # Limited tons of fertilizer (130)

@objective(m, Max, 230w + 190c) # Maximize profit

s = solve(m)
println(s)
println("Acres of wheat: ", getvalue(w))
println("Acres of corn: ", getvalue(c))
println("Optimal (max) profit: ", getobjectivevalue(m))
```

#### Separating Parameters (2b)

```julia
# The types of seed
seeds = [:wheat, :corn]

# Workers required to plant each seed type
workers_needed = Dict(:wheat => 3, :corn => 2)

# Fertilizer needed to plan each seed type (in tons)
fertilizer_needed = Dict(:wheat => 2, :corn => 4)

# Profit made per acre by each seed type ($)
profit = Dict(:wheat => 230, :corn => 190)

# Quantities of workers and fertilizer
num_workers     = 100
num_fertilizer  = 130
num_acres = 45
```

#### Model (2b)

```julia
using JuMP, Clp
m = Model(solver = ClpSolver())

@variable(m, crops[seeds] >= 0) # Dictionary listing how many acres per seed type were planted

@expression(m, total_workers, sum(crops[i] * workers_needed[i] for i in seeds))
@expression(m, total_fertilizer,    sum(crops[i] * fertilizer_needed[i] for i in seeds))
@expression(m, total_profit,  sum(crops[i] * profit[i]  for i in seeds))

@constraint(m, crops[:wheat] + crops[:corn] <= num_acres)  # Total number of acres planted shouldn't exceed 45
@constraint(m, total_workers <= num_workers)               # Total number of workers <= 100
@constraint(m, total_fertilizer <= num_fertilizer)         # Total tons of fertilizer <= 130

@objective(m, Max, total_profit)

s = solve(m)
println(s)
println(getvalue(crops))
println("Optimal (max) profit: ", getvalue(total_profit))
```
