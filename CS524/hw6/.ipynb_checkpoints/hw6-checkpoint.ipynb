#### Part D (Direction of vector)

Let $v = \begin{bmatrix}
  x \\
  y \\
  z
 \end{bmatrix}$

The constraint from part one can be written as $v^{T}Q{v} \leq 1$ so:

$$\begin{bmatrix}
  x \\
  y \\
  z
 \end{bmatrix}^{T}
 \begin{bmatrix}
  2 & 4 & -3 \\
  4 & 2 & -3 \\
  -3 & -3 & 9
 \end{bmatrix}
 \begin{bmatrix}
  x \\
  y \\
  z
 \end{bmatrix} \leq 1$$

Define new coordinates like in class:

$$z = U^{T}v$$

$$v = Uz$$

$$Q = z^{T}\lambda z$$

So since we've redefined $Q$ our new quadratic form can be written:

$$z^{T}\lambda z \leq 1$$

We know that $||v||^2 = ||z||^2$ through simple calculations. So by making $||z||^2$ large, we can equivalently make $||v||^2$ large. So we want a direction to maximize $z$ while adhering to the constraint above.

```julia
# Get the data from CSV
using CSV
raw = CSV.read("lasso_data.csv")
x = raw[1:end,1]
y = raw[1:end,2]
```

```julia
# Order polynomial
k = 5

# Create the A matrix
num_rows = length(x)
A = zeros(num_rows,k+1)
for i = 1:num_rows
    for j = 1:k+1
        A[i,j] = x[i]^(k+1-j)
    end
end

using JuMP, Mosek

m = Model(solver=MosekSolver(LOG=0))

# Vector of coefficients (x)
@variable(m, u[1:k+1])

# Minimize the difference between equation solutions and true solutions
@objective(m, Min, sum((y - A*u).^2))

status = solve(m)
uopt = getvalue(u)
println(status)
println(uopt)
```

```julia
using PyPlot

# Evaluate over new function
xfine = range(0, stop = 0.6, length = 100)
ffine = ones(100)
for j = 1:k
    ffine = [ffine.*xfine ones(100)]
end
yfine = ffine * uopt

# Plot
figure(figsize=(10,5))
plot(x, y, "b.", markersize=5)
plot(xfine, yfine, "r-")
axis([0,0.8,-1.5,1.5])
grid(true)
```

#### Part B (Ridge)

```julia
# Order polynomial
k = 15

# Create the A matrix
num_rows = length(x)
A = zeros(num_rows,k+1)
for i = 1:num_rows
    for j = 1:k+1
        A[i,j] = x[i]^(k+1-j)
    end
end

using JuMP, Mosek

lambda = 1/(10^6)

m = Model(solver=MosekSolver(LOG=0))

# Vector of coefficients (x)
@variable(m, u[1:k+1])

# Minimize the difference between equation solutions and true solutions
@objective(m, Min, sum((y - A*u).^2) + lambda*sum(u.^2))

status = solve(m)
uopt = getvalue(u)
println(status)
println(uopt)
```

```julia
using PyPlot

# Evaluate over new function
xfine = range(0, stop = 0.6, length = 100)
ffine = ones(100)
for j = 1:k
    ffine = [ffine.*xfine ones(100)]
end
yfine = ffine * uopt

# Plot
figure(figsize=(10,5))
plot(x, y, "b.", markersize=5)
plot(xfine, yfine, "r-")
axis([0,0.8,-1.5,1.5])
grid(true)
```

The coefficients have become much smaller.

#### Part C (Lasso)

```julia
# Order polynomial
k = 15

# Create the A matrix
num_rows = length(x)
A = zeros(num_rows,k+1)
for i = 1:num_rows
    for j = 1:k+1
        A[i,j] = x[i]^(k+1-j)
    end
end

using JuMP, Mosek

# Small lambda
lambda = 0.0001

m = Model(solver=MosekSolver(LOG=0))

@variable(m, u[1:k+1])
@variable(m, v[1:k+1])

@constraint(m, u .<= v)
@constraint(m, -v .<= u)

@objective(m, Min, sum((y - A*u).^2) + lambda*sum(v))

status = solve(m)
uopt = getvalue(u)
println(status)
println(uopt)
```

```julia
using PyPlot

# Evaluate over new function
xfine = range(0, stop = 0.6, length = 100)
ffine = ones(100)
for j = 1:k
    ffine = [ffine.*xfine ones(100)]
end
yfine = ffine * uopt

# Plot
figure(figsize=(10,5))
plot(x, y, "b.", markersize=5)
plot(xfine, yfine, "r-")
axis([0,0.8,-1.5,1.5])
grid(true)
```
