### Problem 1

#### Part A (Cubic Fit)

```julia
# Get the data from CSV
using CSV
raw = CSV.read("xy_data.csv")
x = raw[1:end,1]
y = raw[1:end,2]

# Plot it
using PyPlot
figure(figsize=(15,5))
plot(x,y,"b.", markersize=5)
axis([0,10,0,1])
grid(true)
```

```julia
# Order polynomial
k = 3

# Create the A matrix
num_rows = length(x)
A = zeros(num_rows,3)
for i = 1:num_rows
    for j = 1:k
        A[i,j] = x[i]^(k+1-j)
    end
end
```

```julia
using JuMP, Gurobi

m = Model(solver=GurobiSolver(OutputFlag=0,BarHomogeneous=1))

# Vector of coefficients (x)
@variable(m, u[1:k])

# Minimize the difference between equation solutions and true solutions
@objective(m, Min, 1/1000*sum((y - A*u).^2))

status = solve(m)
uopt = getvalue(u)
println(status)
println(uopt)
```

```julia
using PyPlot

# Evaluate over new function
xfine = range(0, stop = 10, length = 100)
ffine = ones(100)
for j = 1:k
    ffine = [ffine.*xfine xfine]
end
ffine = ffine[:, 2:k+1]
yfine = ffine * uopt

# Plot
figure(figsize=(15,5))
plot(x, y, "b.", markersize=5)
plot(xfine, yfine, "r-")
axis([0,10,0,1])
grid(true)
```

#### Part B (Spline Fit)

```julia
# Order polynomial
k = 2

# Get the counts of x data points falling into p or q range
pcount = 0
qcount = 0
for i in 1:length(x)
    if 0<=x[i]<4
        pcount = pcount+1
    end
    if 4<=x[i]<10
       qcount = qcount+1 
    end
end

# Get the solution matrices (b)
pb = y[1:pcount]
qb = y[pcount+1:length(x)]

# Make the A matrix for P function (pcount amount of rows, k+1 cols)
pa = zeros(pcount, k+1)
for i = 1:pcount        # Go over rows
    for j = 1:k+1       # Go over cols
        pa[i,j] = x[i]^(k+1-j) # Fill in
    end
end

# Make the matrix for Q function (qcount amount of rows, k+1 cols)
qa = zeros(qcount, k+1)
for i = 1:qcount                        # Go over rows
    for j= 1:k+1                        # Go over cols
        qa[i,j] = x[i+pcount]^(k+1-j)    # Fill in i+pcount since in order
    end
end
```

```julia
using JuMP, Gurobi

m = Model(solver=GurobiSolver(OutputFlag=0,BarHomogeneous=1))

# Make two x variable vectors (one for each A matrix)
@variable(m, px[1:k+1])
@variable(m, qx[1:k+1])

# Same value and slope at x = 4
@constraint(m, 16px[1] + 4px[2] + px[3] == 16qx[1] + 4qx[2] + qx[3])
# Take the derivative to get slope at x = 4
@constraint(m, 8px[1] + px[2] == 8qx[1] + qx[2])
# Fluorescence zero when intensity is zero
@constraint(m, px[3] == 0)

# Minimize the differences of respective y-Ax
@objective(m, Min, sum((pb - pa*px).^2) + sum((qb - qa*qx).^2))

status = solve(m)
optimalpx = getvalue(px)
optimalqx = getvalue(qx)

println(status)
println(optimalpx)
println(optimalqx)
```

```julia
using PyPlot

# Evaluate p over new range
p_xfine = range(0, stop = 4, length = 40)
p_ffine = ones(40)
for j = 1:k
    p_ffine = [p_ffine.*p_xfine ones(40)]
end
p_yfine = p_ffine * optimalpx

# Evaluate q over new range
q_xfine = range(4, stop = 10, length = 60)
q_ffine = ones(60)
for j = 1:k
    q_ffine = [q_ffine.*q_xfine ones(60)]
end
q_yfine = q_ffine * optimalqx

# Plot
figure(figsize=(15,5))
plot(x, y, "b.", markersize=5)
plot(p_xfine, p_yfine, "r-")
plot(q_xfine, q_yfine, "r-")
axis([0,10,0,1])
grid(true)
```

### Problem 2

#### Part A (MA vs AR Models)

```julia
# Get the data from CSV
using CSV
raw2 = CSV.read("uy_data.csv")
u = raw2[:,1]
y = raw2[:,2]
```

```julia
using LinearAlgebra

# MA Model (similar to class code)
ma_a = zeros(length(u),5)
for i = 1:5
    ma_a[i:end,i] = u[1:end-i+1]
end
wopt = ma_a\y
yhat = ma_a*wopt

figure(figsize=(12,4))
plot(y,"g.-",yhat,"m.-")
legend(["true y", "MA y"], loc="lower right");
title("MA model");
println()
println("Norm: ", LinearAlgebra.norm(yhat-y))
```

```julia
using LinearAlgebra

# AR Model
ar_a = zeros(length(u),5)
for i = 1:5
    ar_a[i+1:end,i] = y[1:end-i]
end
ar_wopt = ar_a\y
ar_yhat = ar_a*ar_wopt

figure(figsize=(12,4))
plot(y,"g.-",ar_yhat,"m.-")
legend(["true y", "AR y"], loc="lower right");
title("AR model");
println()
println("Norm: ",LinearAlgebra.norm(ar_yhat-y))
```

#### Part B (ARMA Model)

```julia
# ARMA Model (just add MA and AR together to get predicted Y)
arma_a = ar_a + ma_a
arma_wopt = arma_a\y
arma_yhat = arma_a*arma_wopt

figure(figsize=(12,4))
plot(y,"g.-",arma_yhat,"m.-")
legend(["true y", "ARMA y"], loc="lower right");
title("ARMA model");
println()
println("Norm: ",LinearAlgebra.norm(arma_yhat-y))
```

#### Part C

```julia
println("MA Norm: ",LinearAlgebra.norm(yhat-y))
println("AR Norm: ",LinearAlgebra.norm(ar_yhat-y))
println("ARMA Norm: ",LinearAlgebra.norm(arma_yhat-y))
```

The norm measures the distance between two vectors. Therefore, a lower norm is better in our case since it means the distance between our Ax vector and our b vector is closer (yhat vs true y). So, the ARMA model came closest to predicting the true y which makes sense since it worked with a larger number of data with which to make predictions (both previous u values and previous y values).
